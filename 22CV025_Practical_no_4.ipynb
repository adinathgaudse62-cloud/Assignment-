```python
# solution
fck = float(input(" Enter the value of characteristic compressive strength: "))

# Experimental Determinations
Gca = float(input("Enter the value of specific gravity of CA: "))
Gfa = float(input("Enter the value of specific gravity of FA: "))
Gc = float(input("Enter the value of specific gravity of Cement: "))
Water_Density = float(input("Enter the value of Water Density: "))
AGG_Size = float(input(" Enter the nominal Size of Aggregate: "))
Nature_of_AGG = input("Nature of Aggregates: ")
Slump = float(input("Enter the value of workability of concrete: "))
Admixture = input("Type of Admixture: ")
Exposure_Condition = input("Exposure Condition: ")
Concreting = input("Type of Concreting: ")
Zone = int(input("Zone: "))

# Target Mean Strength
sigma = {
    10: 3.5, 15: 3.5, 20: 4.0, 25: 4.0, 30: 5.0,
    35: 5.0, 40: 5.0, 45: 5.0, 50: 5.0, 55: 5.0
}
ft = fck + sigma.get(fck, 0) * 1.65
print("Target Mean Strength:", ft, "MPa")

# IS456:2000 values for reinforced concrete (Table 5)
WC_ratio_table = {
    "Mild": 0.55, "Moderate": 0.50, "Severe": 0.45,
    "Very Severe": 0.45, "Extreme": 0.40
}
Min_Cement_Content_table = {
    "Mild": 300, "Moderate": 300, "Severe": 320,
    "Very Severe": 340, "Extreme": 360
}
WC_ratio = WC_ratio_table.get(Exposure_Condition, None)
Min_Cement_Content = Min_Cement_Content_table.get(Exposure_Condition, None)
print("W/C Ratio:", WC_ratio)
print("Minimum Cement Content:", Min_Cement_Content, "kg/m^3")

# Water Content lookup and adjustments
Water_Content_table = {10: 208, 20: 186, 40: 165}
Water_Content = Water_Content_table.get(AGG_Size)
if Water_Content is None:
    raise ValueError("Invalid AGG_Size.")

# Slump adjustment
if Slump == 75:
    Water_Content *= 1.03
elif Slump == 100:
    Water_Content *= 1.06
elif Slump == 125:
    Water_Content *= 1.09
elif Slump == 150:
    Water_Content *= 1.12
elif Slump == 175:
    Water_Content *= 1.15
elif Slump == 200:
    Water_Content *= 1.18

# Nature of aggregate adjustment
if Nature_of_AGG.lower().startswith("sub"):
    Water_Content *= 1.03
elif Nature_of_AGG.lower() == "gravel":
    Water_Content -= 20
elif Nature_of_AGG.lower() == "round":
    Water_Content *= 0.96

# Admixture adjustment
if Admixture.lower() == "plasticizer":
    Water_Content *= 0.9
elif Admixture.lower() == "super-plasticizer":
    Water_Content *= 0.8

print("Water Content:", Water_Content, "kg/m^3")

# Cement Content
Cement_Content = Water_Content / WC_ratio
print("Cement Content:", Cement_Content, "kg/m^3")
if Cement_Content > 450:
    Cement_Content = 450
    print("Adjusted Cement Content to 450 kg/m^3")

# Volumes
Vol_Cement = Cement_Content / (Gc * Water_Density)
Vol_Water = Water_Content / Water_Density
Vol_AGG = 1 - Vol_Cement - Vol_Water
print("Vol Cement:", Vol_Cement, "m^3")
print("Vol Water:", Vol_Water, "m^3")
print("Vol Aggregates (total):", Vol_AGG, "m^3")

# Zone-based fractions
Zone_ID = {
    1: {10: 0.44, 20: 0.60, 40: 0.69},
    2: {10: 0.46, 20: 0.62, 40: 0.71},
    3: {10: 0.48, 20: 0.64, 40: 0.73},
    4: {10: 0.50, 20: 0.66, 40: 0.75}
}
Fraction = Zone_ID.get(Zone, {}).get(AGG_Size, None)
if Fraction is None:
    raise ValueError("Invalid Zone or AGG_Size.")

# Adjust Fraction for W/C ratio
if abs(WC_ratio - 0.45) < 1e-6:
    Fraction *= 1.01
elif abs(WC_ratio - 0.40) < 1e-6:
    Fraction *= 1.02
elif abs(WC_ratio - 0.55) < 1e-6:
    Fraction *= 0.99
elif abs(WC_ratio - 0.60) < 1e-6:
    Fraction *= 0.98

print("Coarse Aggregate fraction:", Fraction)

Vol_CA = Vol_AGG * Fraction
Vol_FA = Vol_AGG - Vol_CA
print("Volume Coarse Agg:", Vol_CA, "m^3")
print("Volume Fine Agg:", Vol_FA, "m^3")

Mass_CA = Vol_CA * Gca * Water_Density
Mass_FA = Vol_FA * Gfa * Water_Density
print("Mass Coarse Agg:", Mass_CA, "kg/m^3")
print("Mass Fine Agg:", Mass_FA, "kg/m^3")

# Batching ratios
print("Weight Batching (Cement : FA : CA : Water)")
print(f"{Cement_Content:.2f} : {Mass_FA:.2f} : {Mass_CA:.2f} : {Water_Content:.2f}")

print("Volume Batching (Cement : FA : CA : Water)")
print(f"{Vol_Cement:.4f} : {Vol_FA:.4f} : {Vol_CA:.4f} : {Vol_Water:.4f}")
```
